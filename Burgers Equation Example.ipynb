```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from PDE_Burgers_Chebyshev_Tau import *
```

```python
N = 20
v = 0.01
t = np.arange(0,10,0.01)
PDE = PDE_Burgers(N,v)
U = PDE.Generate_u(t)
```

```python
t = 50
L = 100

font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 25,
        }

u_func = PDE.build_uN(U[t],N)
X = np.linspace(-1,1,L)
Y = PDE.sample_uN(u_func,L)
plt.figure(figsize=(20,10))
plt.plot(X,Y)
plt.ylabel('u_N value',fontdict = font)
plt.xlabel('x (ranging from -1 to 1)',fontdict = font)
plt.title("Burger Equation",fontdict = font)
plt.show
```

```python
L = 100
step = 5


X = np.linspace(-1,1,L)
font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 25,
        }
def animatep(i):
    t = step*i
    plt.cla()
    plt.xlim([-1, 1])
    plt.ylim([-1, 1])
    u_func = PDE.build_uN(U[t],N)
    Y = PDE.sample_uN(u_func,L)
    
    plt.ylabel('u_N value',fontdict = font)
    plt.xlabel('x (ranging from -1 to 1)',fontdict = font)
    plt.title("Burger Equation",fontdict = font)
    plt.plot(X,Y)
```

```python
#%matplotlib qt
fig = plt.figure(figsize=(12, 6))
ani = FuncAnimation(fig, animatep, frames=200, interval=50, repeat=False)

ani.save('Burgers_Evolution_ChebyShev_Tau.mp4',writer = 'ffmpeg',fps=30)
```
